```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

gpt_score ["Model"] = "GPT-4o-mini"
gemini_score ["Model"] = "Gemini-1.5-Flash"
deepseek_score ["Model"] = "DeepSeek-V3"
claude_score ["Model"] = "Claude-3.5-Sonnet"

df = pd.concat([gpt_score, gemini_score, deepseek_score, claude_score],
               ignore_index=True)
```

```python
df.columns
```

```python
sections = ["Introduction", "Key Factors", "Historical & Theoretical Perspectives",
            "Proposed Resolution Strategies", "Key Takeaways"]

melted = (
    df
    .melt(id_vars=["Model"],  
          value_vars=[f"Score - {s} - Tf-idf" for s in sections] +
                     [f"Score - {s} - DL"     for s in sections] +
                     [f"Score - {s} - USE"    for s in sections] +
                     [f"Score - {s} - BLEU"   for s in sections],
          var_name="MetricSection", value_name="Score")
    .assign(          
        Metric=lambda d: d.MetricSection.str.extract(r"- ([A-Za-z-]+)$", expand=False),
        Section=lambda d: d.MetricSection.str.extract(r"Score - (.*?) -",  expand=False))
)
```

```python
plt.figure(figsize=(11,4))
sns.barplot(data=melted, x="Section", y="Score", hue="Model", ci=None)
plt.ylim(0,1); plt.ylabel("Mean score"); plt.xticks(rotation=15); plt.title("Section quality by model");
```

```python
heat = (melted.groupby(["Model","Metric"])["Score"]
               .mean()
               .unstack())
sns.heatmap(heat, annot=True, vmin=0, vmax=1, fmt=".2f", cmap="viridis")
plt.title("Average score by metric"); plt.ylabel(""); plt.xlabel("");
```

```python
sns.pairplot(df[[c for c in df.columns if "Total Score" in c]+["Final Score"]], kind="reg")
```

```python
import numpy as np

def radar(model_name):
    means = (melted[melted.Model==model_name]
             .groupby("Metric")["Score"].mean().reindex(["Tf-idf","DL","USE","BLEU"]))
    labels = means.index.tolist()
    values = means.values
    angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)
    fig=plt.figure(figsize=(4,4)); ax=fig.add_subplot(111, polar=True)
    ax.plot(np.append(angles, angles[0]), np.append(values, values[0]), linewidth=2)
    ax.fill(np.append(angles, angles[0]), np.append(values, values[0]), alpha=0.25)
    ax.set_thetagrids(angles*180/np.pi, labels); ax.set_ylim(0,1); ax.set_title(model_name)
    plt.show()

for m in df.Model.unique(): radar(m)
```

```python
m1, m2 = "Total Score USE", "Total Score DL"
avg  = df[[m1,m2]].mean(1)
diff = df[m1] - df[m2]

plt.axhline(diff.mean(), ls="--"); plt.scatter(avg, diff)
plt.xlabel("Mean of USE & DL"); plt.ylabel("USE − DL"); plt.title("Bland–Altman");
```

```python
def radar_metric(metric):
    vals = (melted[melted.Metric==metric]
            .groupby("Section")["Score"]
            .mean().reindex(sections))
    angles = np.linspace(0,2*np.pi,len(sections),endpoint=False)
    fig,ax=plt.subplots(subplot_kw=dict(polar=True))
    ax.plot(np.r_[angles,angles[0]], np.r_[vals,vals.iloc[0]], lw=2)
    ax.fill(np.r_[angles,angles[0]], np.r_[vals,vals.iloc[0]], alpha=.25)
    ax.set_thetagrids(angles*180/np.pi, sections); ax.set_ylim(0,1)
    ax.set_title(f"{metric} strengths by section")
    plt.show()

for metric in ["Tf-idf","DL","USE","BLEU"]: radar_metric(metric)
```
